```python
import pandas as pd 
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
# as the input value has categorical data in it, convert in to numeric with get dummies
dataset=pd.get_dummies(dataset, drop_first=True)
```

```python
dataset
```

```python
#Split input and output 
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'charges', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
#now split the Train set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#create the model
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson',splitter='best')
regressor=regressor.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#Evaluation metrics
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
